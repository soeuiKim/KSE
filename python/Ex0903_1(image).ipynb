```python
import cv2
import matplotlib.pyplot as plt

img1_ = cv2.imread('C:\KSE\examApp/image1.jpg') # 이미지파일로드

img1 = cv2.cvtColor(img1_, cv2.COLOR_BGR2RGB) #bgr형식을 rgb형식으로

plt.axis('off') #축눈금제거
plt.imshow(img1) #이미지변수를 이용하여 화면 출력
plt.show()
```

```python
img1_ = cv2.imread('C:\KSE\examApp/image1.jpg') # 이미지파일로드
img1 = cv2.cvtColor(img1_, cv2.COLOR_BGR2RGB) #bgr형식을 rgb형식으로

img2_ = cv2.imread('C:\KSE\examApp/image2.jpg') # 이미지파일로드
img2 = cv2.cvtColor(img2_, cv2.COLOR_BGR2RGB) #bgr형식을 rgb형식으로

plt.subplot(121)
plt.axis('off')
plt.imshow(img1)

plt.subplot(122)
plt.axis('off')
plt.imshow(img2)

plt.show()
```

```python
img1_ = cv2.imread('C:\KSE\examApp/image1.jpg') 
img1 = cv2.cvtColor(img1_, cv2.COLOR_BGR2RGB)

font = cv2.FONT_HERSHEY_SIMPLEX
num = 100
numstr = 'Num: %d' % (num)
# 메모리, 문자열, 좌표, 폰트, 크기, 색상, 굵기
cv2.putText(img1_, numstr, (80,120), font, 2, (255,255,255), 5)
cv2.putText(img1_, 'cat', (450,400), font, 2, (255,255,255), 5)
cv2.putText(img1_, 'wow', (img1.shape[1] - 150,img1.shape[0] - 100), font, 2, (0,0,255), 5)

cv2.imwrite('C:\KSE\examApp/image1_new.jpg', img1_) #메모리 영상을 파일로 저장


plt.axis('off')
plt.imshow(img1_)
plt.show()
```
