```python
import pandas as pd
import glob # 경로와 이름을 지정하여 파일을 처리
import re # 메타문자를 이용하여 특정 규칙을 작성하는 정규식 사용모듈
from functools import reduce # 2차원리스트를 1차원리스트로
from nltk.tokenize import word_tokenize # 단어토큰화작업
from nltk.corpus import stopwords # 불용어정보제공
from nltk.stem import WordNetLemmatizer # 표제어 추출
from collections import Counter # 데이터집합개수  계산
import matplotlib.pyplot as plt 
from wordcloud import STOPWORDS, WordCloud # 워드클라우드그리기(불용어모듈,워드클라우드모듈)
```

```python
all_files = glob.glob('data/myCabinetExcelData*.xls')
all_files
```

```python
all_files_data = []
for file in all_files:
    # 리스트에 있는 파일 이름 이용, 엑셀 파일 읽기
    data_frame = pd.read_excel(file)
    # 파일 내용을 all_files_data에 추가
    all_files_data.append(data_frame)
all_files_data[0]
```

```python
# all_files_data를 세로축 기준으로 병합, all_files_data_concat 리스트에 저장  (axis = 1 : 가로축기준으로 저장)
all_files_data_concat = pd.concat(all_files_data, axis=0, ignore_index = True)
all_files_data_concat
```

```python
# 병합된 all_files_data_concat을 cvs파일로 저장
all_files_data_concat.to_csv('data/riss_bigdata.csv',index=False)
```

```python
all_files_data_concat.head()
```

```python
# 데이터 전처리
# 제목컬럼확인
all_title = all_files_data_concat['제목']
all_title
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
# 전처리 진행 : 
# 1. 영어가 아닌 단어 제거하고 소문자로 정규화 -> 단어 토큰화
# 2. 불용어 제거하고 단어형태 일반화를 위해 표제어 추출
stopWords = set(stopwords.words("english")) # 영어불용어
lemma = WordNetLemmatizer() # 표제어추출 객체생성
words = []
# all_title 제목에대해 정규시긍로 만든 규칙 적용
# 영어가 아닌 것, 즉 알파벳으로 시작하지 않는 단어는 공백으로 치환하여 제거하고
# 소문자로 정규화하고
# 단어 토큰화(words_tokenize) 한다.
# 그리고 불용어를 제가한 후에 표제어 추출(lemmaize)한다.
for title in all_title:
    EnWords = re.sub(r"[^a-zA-Z]+", " ", str(title))
    EnWordsToken = word_tokenize(EnWords.lower())
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopWords]
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop]
    words.append(EnWordsTokenStopLemma)
print(words)
```

```python
# 2차원리스트를 1차원리스트로 변환
words2 = list(reduce(lambda x,y:x+y, words))
print(words2)
```

```python
# 데이터 탐색, 분석 모델 구축
# 단어별로 출현 횟수를 계산하여 딕셔너리 count를 생성
count = Counter(words2)
count
```

```python
# 출현횟수가 많은 상위 50개 단어 중에서 단어 길이가 1보다 큰것만 저장

word_count = dict()

for tag, counts in count.most_common(50):
    if(len(str(tag))>1):
        word_count[tag] = counts
        print("%s : %d" % (tag, counts))
```

```python
#
sorted_Keys = sorted(word_count, key = word_count.get, reverse = True)
sorted_Values = sorted(word_count.values(), reverse = True)
plt.bar(range(len(word_count)), sorted_Values, align = 'center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation = '85')
plt.show()
```

```python
all_files_data_concat['doc_count'] = 0
summary_year = all_files_data_concat.groupby('출판일',as_index=False)['doc_count'].count()
summary_year
```

```python
plt.figure(figsize = (12, 5))
plt.xlabel("year")
plt.ylabel("doc-count")
plt.grid(True)
plt.plot(range(len(summary_year)), summary_year['doc_count'])
plt.xticks(range(len(summary_year)), [text for text in summary_year['출판일']])
plt.show()
```

```python
stopwords = set(STOPWORDS)
wc = WordCloud(background_color = 'black', stopwords = stopwords, width = 800, height = 600)
cloud = wc.generate_from_frequencies(word_count)
plt.figure(figsize = (8,8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
